```python
##This actually looks like the quotient. Eventually it should give remainer though?
def rem(n,k):
	q=0
	r=0
	if(n>k):
		if(n%k==0):
			q=n//k-1
			r=k
		else:
			q=n//k
			r=n%k
	else:
		q=0
	return(q)
```

```python
##Converts numbers to words. Default is English alphabet.
def numToString(n):
	k=26
	w=''
	a=n+1
	while(a>k):
		j=0
		q=a
		while(q>k):
			q=rem(q,k)
			j=j+1
		w=w+LETTERS[q-1]
		a=a-q*k**j
	w=w+LETTERS[a-1]
	return(w)
```

```python
#Finds the number in the dictionary associated with string n
def stringToNum(n):
		k=26
		x=0
		for i in range(1,len(n)+1):
			x=x+(LETTERS.index(n[len(n)-(i-1)-1])+1)*k**(i-1)
		return(x-1)
```

```python
#Prints all words from m to n. Display has w words in each row.
def printWindow(m,n,w):
    i=j=k=0
    sk=""
    q = (n-m)//w
    for j in range(q):
        for k in range(w):
            sk+= numToString(m+j*w+k)+" "
        print(sk)
        sk = ""
    for k in range(m+q*w,n+1):
        sk+= numToString(k)+" "
    print(sk)
```

```python
#Prints word and all words within a radius ir of word. w words are listed in each row.
def printWordWindow(word,ir,w):
    printWindow(stringToNum(word)-ir,stringToNum(word)+ir,w)
```

# For arbitrary alphabets

```python
#Returns string of number n in alphabet alph.
def numToString(n,alph = LETTERS):
	k=len(alph)
	w=''
	a=n+1
	while(a>k):
		j=0
		q=a
		while(q>k):
			q=rem(q,k)
			j=j+1
		w=w+alph[q-1]
		a=a-q*k**j
	w=w+alph[a-1]
	return(w)
```

```python
#Returns number of string n in alphabet alph.
def stringToNum(n,alph = LETTERS):
		k=len(alph)
		x=0
		for i in range(1,len(n)+1):
			x=x+(alph.index(n[len(n)-(i-1)-1])+1)*k**(i-1)
		return(x-1)
```

```python
#Prints all words from m to n in alphabet alph. Each row has w words.
def printWindow(m,n,w,alph = LETTERS):
    i=j=k=0
    sk=""
    q = (n-m)//w
    for j in range(q):
        for k in range(w):
            sk+= numToString(m+j*w+k,alph)+" "
        print(sk)
        sk = ""
    for k in range(m+q*w,n+1):
        sk+= numToString(k,alph)+" "
    print(sk)
```

```python
#Prints word and radius words above and below word in alphabet alph. w words are printed in alphabet alph.
def printWordWindow(word,radius,w,alph = LETTERS):
    center = stringToNum(word,alph)
    printWindow(center-radius,center+radius,w,alph)
```
